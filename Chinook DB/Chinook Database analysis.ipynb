```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('Best_employee.csv')
df1
```

```python
#, dpi=80
plt.figure(figsize=(7, 7))
sns.set_theme(style="whitegrid")
sns.barplot(x=df1['employee_Fname'], y=df1['total_purchases'], data=df1)
plt.xlabel('Employee Name')
plt.ylabel('Total purchases')
plt.title('Best Employee based on total purchases');
```

```python
df2 = pd.read_csv('Famous_artist_city_2.csv')
df2
```

```python
plt.figure(figsize=(10, 7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='city_name', y='max_n_times', hue='artist_name', data=df2, dodge=False)
ax.legend(loc='upper right',prop={'size':12})
plt.title('Top 10 artist listened in these cities')
plt.xlabel('Cities')
plt.ylabel('# times listened')
plt.xticks(rotation=45, fontsize=12)
plt.show();
```

```python
df4 = pd.read_csv('Track_with_format.csv')
#getting the top 10
df4_c = df4[:10].copy()
```

```python
df4_c
```

```python
normalized_col =[]
mi = np.min(df4_c['length'])
mx = np.max(df4_c['length'])
for i in df4_c['length']:
    normalized_col.append((i - mi)/(mx - mi))
```

```python
plt.figure(figsize=(9, 7))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=df4_c['track_name'], y=normalized_col, data=df4_c)
plt.xticks(rotation=70, fontsize=12)
plt.xlabel('Track')
plt.ylabel('Normalized average lenght')
plt.title("longer than average track and in MPEG audio format")
plt.show()
```
